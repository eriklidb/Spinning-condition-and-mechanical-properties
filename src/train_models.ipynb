```python
from model_trainer import *
from dataset import Dataset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap

sns.set_theme(style="white")
seed = 42
```

```python
ds = {}
ds['A'] = Dataset('spinning_data.csv')
ds['B'] = Dataset('spinning_data_embeddings.csv')
cv_type = 'kfold'
n_folds = 3
n_inner_folds = 5
mt = ModelTrainer(n_outer_folds=n_folds, n_inner_folds=n_inner_folds, cv_type=cv_type, random_state=seed)
models = {}
study_params = {
    'n_trails': 50,
    'timeout': 1800,
    'n_jobs': -1
}
```

```python
k = 'A'

models[k] = mt.hyperparameter_search(ds[k], study_name=k, **study_params)
for i, model in enumerate(models[k]):
    save_model(model, f'model_{k}_fold_{i}')

#print(f"\nNested CV RMSEs: {scores[k]}")
#print(f"Mean RMSE: {np.mean(scores[k]):.4f} ± {np.std(scores[k]):.4f}")
```

```python
k = 'A'
models[k] = [None] * n_folds
for i in range(n_folds):
    models[k][i] = load_model(f'model_{k}_fold_{i}')
    models[k][i].estimator.categorical_features = ds[k].categorical_columns
```

```python
k = 'B'
models[k] = [None] * n_folds
for i in range(n_folds):
    models[k][i] = load_model(f'model_{k}_fold_{i}')
    models[k][i].estimator.categorical_features = ds[k].categorical_columns
```

```python
metrics = {}
metrics_mean = {}
y_true = {}
y_pred = {}
y_true_tot = {}
y_pred_tot = {}
for k in 'A', 'B':
    X, Y = ds[k]()
    metrics[k] = [None] * n_folds
    y_true[k] = [None] * n_folds
    y_pred[k] = [None] * n_folds
    for fold, (train_idx, test_idx) in enumerate(mt._outer_cv.split(X)):
        y_true[k][fold] = Y.iloc[test_idx]
        y_pred[k][fold] = models[k][fold].predict(X.iloc[test_idx])
        metrics[k][fold] = compute_metrics(y_true[k][fold], y_pred[k][fold])
    y_true_tot[k] = np.vstack(y_true[k])
    y_pred_tot[k] = np.vstack(y_pred[k])
    metrics_mean[k] = pd.DataFrame(np.array(metrics[k]).mean(axis=0), index=metrics[k][0].index, columns=metrics[k][0].columns)
```

```python
print(metrics_mean['A'].to_string())
print(metrics_mean['B'].to_string())
```

```python
for k in 'A', 'B':
    fig, axs = plt.subplots(5, 1)
    fig.set_figheight(28)
    fig.set_figwidth(7)
    for i in range(5):
        col = ds[k].target_columns[i]
        x = y_pred_tot[k][:,i]
        y = y_true_tot[k][:,i]
        z = np.polyfit(x, y, 1)
        y_hat = np.poly1d(z)(x)
        axs[i].set_title(f'{col}, model {k}')
        axs[i].set_xlabel('Standardized predicted value')
        axs[i].set_ylabel('Standardized actual value')
        axs[i].scatter(x, y, c='b', alpha=.25)
        axs[i].plot(x, y_hat, "r-", lw=1)
        text = r"$y={:0.3f}\;x+{:0.3f}$".format(z[0], z[1]) + "\n" + r"$R^2 = {:0.3f}$".format(metrics_mean[k].loc['R2', col])
        axs[i].text(0.05, 0.95, text, transform=axs[i].transAxes,
        fontsize=14, verticalalignment='top')
    fig.savefig(f'../figures/scatter_{k}.png')
    plt.show()
```

```python
for k in 'A', 'B':
    fig, axs = plt.subplots(5, 1)
    fig.set_figheight(28)
    fig.set_figwidth(7)
    for i in range(5):
        col = ds[k].target_columns[i]
        x = y_pred_tot[k][:,i]
        y = y_true_tot[k][:,i] - y_pred_tot[k][:,i]
        z = np.polyfit(x, y, 1)
        y_hat = np.poly1d(z)(x)
        axs[i].set_title(f'{col}, model {k}')
        axs[i].set_xlabel('Standardized residual')
        axs[i].set_ylabel('Standardized actual value')
        axs[i].scatter(x, y, c='b', alpha=.25)
        axs[i].plot(x, y_hat, "r-", lw=1)
        text = r"$y={:0.3f}\;x+{:0.3f}$".format(z[0], z[1]) + "\n" + r"$R^2 = {:0.3f}$".format(metrics_mean[k].loc['R2', col])
        axs[i].text(0.05, 0.95, text, transform=axs[i].transAxes,
        fontsize=14, verticalalignment='top')
    fig.savefig(f'../figures/residual_{k}.png')
    plt.show()
```

```python
k = 'B'
fold = 0
# Your trained MultiOutputRegressor model
multi_model = models[k][fold]  # MultiOutputRegressor(HistGradientBoostingRegressor(...))
X_sample = ds[k].features # Or a subset like X_train[:200] for faster results

# Copy your input dataframe
X_encoded = X_sample.copy()

# Convert your categorical columns to 'category' dtype
for col in ds[k].categorical_columns:  # list of your categorical column names
    X_encoded[col] = X_encoded[col].astype("category")
    X_encoded[col] = X_encoded[col].cat.codes  # convert to integer codes

# Loop over each target output
for i, estimator in enumerate(multi_model.estimators_):
    #print(f"\n🔹 SHAP values for target: {dataset.columns[-5+i]}")
    
    # Create a SHAP explainer (TreeExplainer is efficient for gradient boosting models)
    explainer = shap.Explainer(estimator)
    
    # Compute SHAP values for the input data
    shap_values = explainer(X_encoded)
    
    # Plot summary plot
    shap.summary_plot(shap_values, features=X_encoded, feature_names=X_encoded.columns, show=False)
    plt.title(f"SHAP values for target: {ds[k].columns[-5+i]}")
    plt.show()
```
